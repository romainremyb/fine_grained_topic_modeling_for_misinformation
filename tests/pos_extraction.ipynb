## AJD-NOUN retrieval NLTK

```python
def add_JJ_NN_pairs(tokens):
    adj_noun_pairs = []
    pos_tagged = pos_tag(tokens)
    for i in range(len(pos_tagged) - 1):
        word, pos = pos_tagged[i]
        next_word, next_pos = pos_tagged[i + 1]
        if pos.startswith('JJ') and next_pos.startswith('NN'):
            adj_noun_pairs.append(word +' '+ next_word)
    return adj_noun_pairs

ex = 'This some hopefully good tests, it is a pythonic way to retrieve pos pairs correctly'
ex = nltk.tokenize.RegexpTokenizer(r'\w+').tokenize(ex.lower())
print(add_JJ_NN_pairs(ex))
for i, j in pos_tag(ex):
    print(i,j)
```
